```python
import pandas as pd
import re
from geopy.geocoders import GoogleV3
import creds
```

```python
df = pd.read_csv('Output\Mid_01_Cleaned.csv')
```

```python
df['landprice'] = df['price']/df['land']
```

```python
df.describe()
```

```python
df['landprice'] = df['landprice'].astype(int)
```

```python
df
```

```python
locs = pd.DataFrame(df.location.unique())
locs.columns = ['location']
```

```python
locs
```

```python
geolocator = GoogleV3(api_key=creds.api_key)
```

```python
locs["loc"] = locs["location"].apply(geolocator.geocode)
```

```python
locs["point"]= locs["loc"].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
locs[['lat', 'lon', 'altitude']] = pd.DataFrame(locs['point'].to_list(), index=locs.index)
```

```python
locs
```

```python
locs = locs.iloc[:,0:5]
```

```python
locs
```

```python
locs.to_csv('Output\Mid_02_Location-Key.csv', index=False)
```

```python
area_key = {}
for i in range(0,55):
    area_key.setdefault(locs.iloc[i,0], []).append(str(locs.iloc[0,1]))
    area_key.setdefault(locs.iloc[i,0], []).append(locs.iloc[i,2])
    area_key.setdefault(locs.iloc[i,0], []).append(locs.iloc[i,3])
    area_key.setdefault(locs.iloc[i,0], []).append(locs.iloc[i,4])
```

```python
for row in range(0, 1928):
    area_name = df.loc[row, 'location']
    df.loc[row, 'loc'] = area_key[area_name][0]
    df.loc[row, 'point'] = str(area_key[area_name][1])
    df.loc[row, 'lat'] = area_key[area_name][2]
    df.loc[row, 'lon'] = area_key[area_name][3]
```

```python
for row in range(0, 1928):
    target = df.loc[row, 'loc']
    regency = ['Sleman Regency', 'Yogyakarta City', 'Bantul Regency', 'Gunung Kidul Regency', 'Kulon Progo Regency']
    df.loc[row, 'regency'] = re.findall(r"(?=("+'|'.join(regency)+r"))", target)[0]
```
