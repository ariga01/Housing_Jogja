```python
import pandas as pd
```

```python
df = pd.concat([
    pd.read_csv('Output\Early_Data_Scraped-Raw.csv'),
    pd.read_csv('Output\Early_Data_Scraped-Raw_Backup_01.csv'),
    pd.read_csv('Output\Early_Data_Scraped-Raw_Backup_02.csv')
    ])
df = df.dropna()
df = df.drop_duplicates(subset=('title','price','land','building','location'), keep='last')
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.dtypes.info
```

```python
df = df.astype(str)
```

```python
cols = ["price","land", "building", "bed", "bath", "floor", "certificate"]
for col in cols :
    df = df[df[f"{col}"].str.contains("&gt;10") == False]
```

```python
df = df[df['location'].str.contains("Yogyakarta") == True]
```

```python
conv = {
    'land': float, 
    'building': float,
    'bed': float,
    'bath':float,
    'floor': float
    }

df = df.astype(conv)
```

```python
conv = {
    'land': int, 
    'building': int,
    'bed': int,
    'bath':int,
    'floor': int
    }

df = df.astype(conv)
```

```python
df = df[(df['land']>0) == True]
```

```python
df
```

```python
df.price = df.price.str.replace(r'\D+', '', regex=True)
df.price = df.price.astype('float')
df.price = df.price.astype('int64')
```

```python
df.reset_index(inplace=True, drop=True)
```

```python
df
```

```python
df.to_csv('Output\Mid_01_Data_Cleaned.csv', index=False)
```
